```python
"""Program to find digital sum of a given Number
   ex: n=123  Digital sum----->1+2+3=6
"""
def getsum(n):
    sum = 0
    while(n != 0):
        sum = sum + (n % 10)
        n = n//10
    return sum;

n = int(input("enter the number: "))
print(getsum(n))
```

```python
#print the following pattern

n = int(input("enter the number"))
for i in range(0,n):
    for j in range(0,i+1):
        print("*",end=' ')
    print(" ")
for i in range(n + 1, 0, -1):  
    for j in range(0, i - 1):  
        print("*", end=' ')  
    print(" ")  
    
```

```python
# Program to reverse a given Number.    ex: n=123   Reversed no is 321
num = int(input("enter the number"))
reverse = 0
while(num != 0):
    sum = num % 10
    reverse =  reverse * 10 + sum
    num = num//10
print(reverse)
    
```

```python
# Program to check whether a given number is a palindrome or not
n = int(input("enter the number: "))
sum = 0
temp = n
while(n != 0):
    r = n % 10
    sum = sum * 10 + r
    n = n//10
if temp == sum:
    print("it's a palindrome number")
else:
    print("it's not a palindrome number")
    
    
```

```python
# Program to check whether a given number is an Armstrong number or not.
n = int(input("enter the number: "))
temp = n
sum = 0
while(n != 0):
    r = n % 10
    sum = sum + r**3
    n = n//10
if temp == sum:
    print("it is a armstrong number")
else:
    print("it's not a armstrong number")
    
```

```python
# Write a Python function  to check whether the given integer is a prime number or not.
num = int(input("enter the number"))
if num > 1:
    for i in range(2, num):
        if(num%i==0):
            print("it's not a prime number")
            break
    else:
            print("it's a prime number")
```

```python
# Write a Python function that checks whether a passed interger is armstrong or not.
n = int(input("enter the number: "))
def armstrong(n):
    temp = n
    sum = 0
    while(n != 0):
        r = n % 10
        sum = sum + r**3
        n = n//10
    if temp == sum:
        print("it is a armstrong number")
    else:
        print("it's not a armstrong number")
print(armstrong(n))    
```

```python
# Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)
n = int(input("enter the number"))
print(factorial(n))
    
```

```python

def even(str):
    str = input("please enter the string: ")
    result = ""
    for i in range(len(str)):
        if i % 2 == 0:
            result = result + str[i]
    return result

print(even(str))
```
